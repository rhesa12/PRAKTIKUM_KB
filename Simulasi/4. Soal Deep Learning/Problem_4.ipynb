## Import Library & Read Data

```python
import pandas as pd
import tensorflow as tf
import numpy as np
import sklearn
```

```python
# JANGAN MENGUBAH KODE INI
df = pd.read_csv('milknew_train.csv')
df
```

<div class="alert alert-block alert-success">
    <h3><b>A.</b> Split Dataset</h3>
</div>

```python
X =  df.drop('Grade', axis=1)#KODE KALIAN DISINI
y =  df['Grade']#KODE KALIAN DISINI

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)#KODE KALIAN DISINI
```

<div class="alert alert-block alert-success">
    <h3><b>B.</b> Preprocessing</h3>
    Encode label dan tahapan preprocessing lain jika perlu
</div>

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()# KODE KALIAN DISINI

# JANGAN MENGUBAH KODE INI
y_train = encoder.fit_transform(y_train)
y_test = encoder.transform(y_test)
```

<div class="alert alert-block alert-success">
    <h3><b>C.</b> Modelling</h3>
</div>

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(64, activation='relu', input_shape=(7,)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(256, activation='relu'),
    # tf.keras.layers.Dense(128, activation='relu'),
    # tf.keras.layers.Dense(256, activation='relu'),
    tf.keras.layers.Dense(3, activation='softmax'),
    #KODE KALIAN DISINI
])

model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
    #KODE KALIAN DISINI
)

history = model.fit(
    X_train, y_train, epochs=500, validation_split=0.2
    #KODE KALIAN DISINI
)
```

<div class="alert alert-block alert-success">
    <h3><b>D.</b> Model Evaluation</h3>
</div>

```python
# JANGAN MENGUBAH KODE INI
model.evaluate(X_test, y_test)
```

<div class="alert alert-block alert-success">
    <h3><b>E.</b> Save Prediction</h3>
    Jangan lupa untuk menerapkan preprocessing yang sama jika melakukan preprocessing sebelumnya
</div>

```python
df_test = pd.read_csv('milknew_test.csv')

#KODE KALIAN DISINI (JIKA MELAKUKAN PREPROCESSING SEBELUMNYA)

# JANGAN MENGUBAH KODE INI
y_pred = model.predict(df_test)
y_pred = np.argmax(y_pred, axis=1)
y_pred_decoded = encoder.inverse_transform(y_pred)
np.save('y_pred', y_pred_decoded)
```
