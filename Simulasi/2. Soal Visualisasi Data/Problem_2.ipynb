## Import Library & Read Data

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Housing.csv')
df.head()
```

<div class="alert alert-block alert-success">
    <h3><b>2.1</b> Carilah salah dua atribut yang memiliki korelasi linear, kemudian visualisasikan korelasi kedua atribut tersebut</h3>
</div>

```python
#KODE KALIAN DISINI
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
sns.scatterplot(df, x='price', y='area')
plt.show()
```

**KESIMPULAN KALIAN DISINI**
<hr>


<div class="alert alert-block alert-success">
    <h3><b>2.2</b> Bagaimana penyebaran nilai dari harga rumah</h3>
</div>

```python
#KODE KALIAN DISINI
from sklearn.preprocessing import MinMaxScaler
ss = MinMaxScaler()

price = ss.fit_transform(df[['price']])
sns.histplot(price)
plt.show()
```

```python
df[['price']] = price
sns.scatterplot(df, x='price', y='area')
plt.show()
```

**KESIMPULAN KALIAN DISINI**
<hr>


<div class="alert alert-block alert-success">
    <h3><b>2.3</b> Visualisasikan korelasi antar atribut</h3>
</div>

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

**KESIMPULAN KALIAN DISINI**
<hr>


<div class="alert alert-block alert-success">
    <h3><b>2.4</b> Tampilkan perbandingan persentase rumah pada atribut pemanas air, pendingin ruangan, basement, dan kapasitas parkir</h3>
</div>

```python
#KODE KALIAN DISINI
plt.figure(figsize=(15, 5))
plt.subplot(1, 4, 1)
plt.pie(df['hotwaterheating'].value_counts(), labels=df['hotwaterheating'].unique(), autopct='%.2f%%')
plt.title('Pemanas Air')
plt.subplot(1, 4, 2)
plt.pie(df['airconditioning'].value_counts(), labels=df['airconditioning'].unique(), autopct='%.2f%%')
plt.title('Pendingin Ruangan')
plt.subplot(1, 4, 3)
plt.pie(df['basement'].value_counts(), labels=df['basement'].unique(), autopct='%.2f%%')
plt.title('Basement')
plt.subplot(1, 4, 4)
plt.pie(df['parking'].value_counts(), labels=df['parking'].unique(), autopct='%.2f%%')
plt.title('Parkiran')

plt.show()
```

**KESIMPULAN KALIAN DISINI**
<hr>


<div class="alert alert-block alert-success">
    <h3><b>2.5</b> Identifikasi outlier yang terdapat pada dataset menggunakan visualisasi grafik</h3>
</div>

```python
#KODE KALIAN DISINI
plt.figure(figsize=(15, 5))
plt.subplot(1, 6, 1)
plt.boxplot(df['price'])
plt.title('harga')
plt.subplot(1, 6, 2)
plt.boxplot(df['area'])
plt.title('area')
plt.subplot(1, 6, 3)
plt.boxplot(df['bedrooms'])
plt.title('kamar')
plt.subplot(1, 6, 4)
plt.boxplot(df['bathrooms'])
plt.title('wc')
plt.subplot(1, 6, 5)
plt.boxplot(df['stories'])
plt.title('cerita')
plt.subplot(1, 6, 6)
plt.boxplot(df['parking'])
plt.title('parkiran')
plt.show()
```
