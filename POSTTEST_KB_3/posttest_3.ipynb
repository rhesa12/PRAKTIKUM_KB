## Import Library ##

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Melihat Data Yang Cocok ##

```python
pd.read_csv('smoking.csv')
```

```python
smoke = pd.read_csv('smoking.csv')
```

```python
smoke.drop(['Unnamed: 0'], axis=1)
data = smoke.groupby('marital_status').mean()
data
```

## Presentase perkawinan perokok dan non perokok ##

```python
data = smoke[smoke['smoke'] == 'Yes']['marital_status'].value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%",  startangle=90)
plt.title('data status perkawinan perokok ')
plt.show()
```

```python
data = smoke[smoke['smoke'] == 'No']['marital_status'].value_counts()

plt.pie(data, labels=data.index, autopct="%1.1f%%",  startangle=90)
plt.title('data status perkawinan bukan perokok ')
plt.show()
```

## kolerasi table ##

```python
cor_matrix = smoke.iloc[0:, 1:].corr()

sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```
